```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image_path="C:/Users/User/OneDrive/Desktop/github/edge detector/download.jpeg"
img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("gray",gray)
```

```python

input_image_gray = np.float64(gray)

# Display Input Image
plt.figure(figsize=(6, 6))
plt.imshow(input_image_gray, cmap='gray')
plt.title('Input Image')
plt.show()

```

```python
# Sobel Operator Masks
Mx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
My = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

```

```python
# Pre-allocate filtered_image array with zeros
filtered_image = np.zeros(input_image_gray.shape)

# Edge Detection code
for i in range(input_image_gray.shape[0] - 2):
    for j in range(input_image_gray.shape[1] - 2):
        # Gradient x and y
        Gx = np.sum(Mx * input_image_gray[i:i+3, j:j+3])
        Gy = np.sum(My * input_image_gray[i:i+3, j:j+3])

        # Calculate magnitude
        filtered_image[i+1, j+1] = np.sqrt(Gx**2 + Gy**2)

# Normalize filtered image
filtered_image_normalized = np.uint8(filtered_image / filtered_image.max() * 255)
print(filtered_image.max())


# show filtered image
plt.figure(figsize=(6, 6))
plt.imshow(filtered_image_normalized, cmap='gray')
plt.title('Filtered Normalised Image')
plt.show()

```

```python
# threshold value set
thresholdValue = 100  # varies between [0 255]
_, output_image = cv2.threshold(filtered_image_normalized, thresholdValue, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(6, 6))
plt.imshow(output_image, cmap='gray')
plt.title('Edge Detected Image')
plt.show()
```
